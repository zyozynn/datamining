```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from PIL import Image
```

```python
# Load grayscale image
img_gray = cv2.imread('Cat.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Calculate histogram
hist_gray, bins_gray = np.histogram(img_gray.flatten(), 256, [0,256])
```

```python
# Plot histogram
plt.hist(img_gray.flatten(), 256, [0,256], color='gray')
plt.xlim([0,256])
plt.xlabel('Gray Level')
plt.ylabel('Pixel Count')
plt.title('Gray Scale Histogram')
plt.show()
```

```python
# Histogram Equalization
img_eq = cv2.equalizeHist(img_gray)
```

```python
# Calculate histogram after equalization
hist_eq, bins_eq = np.histogram(img_eq.flatten(), 256, [0,256])
```

```python
# Plot histogram after equalization
plt.hist(img_eq.flatten(), 256, [0,256], color='gray')
plt.xlim([0,256])
plt.xlabel('Gray Level')
plt.ylabel('Pixel Count')
plt.title('Histogram Equalization')
plt.show()
```

```python
# Histogram RGB
cf = Image.open(r'D:\Wallpaper\Tiny Tan.jpg')
lbr, tgi = (cf.size)

bakir = []
bakig = []
bakib = []

for x in range (lbr):
    for y in range(tgi):
        r,g,b = cf.getpixel((x,y))
        bakir.append(r)
        bakig.append(g)
        bakib.append(b)
        
jb = 255
fig, axs = plt.subplots(1,3,figsize=(15,5))
# axs[0].imshow(cf)
axs[0].hist(bakir, jb, facecolor='red', alpha=0.5)
axs[1].hist(bakig, jb, facecolor='green', alpha=0.5)
axs[2].hist(bakib, jb, facecolor='blue', alpha=0.5)
plt.show()
```
