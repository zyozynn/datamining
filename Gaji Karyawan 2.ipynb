```python
# Program Program Gaji Karyawan by Siti Ainun Rodhiyah
# TI-2B - 44.33.21.1.24 

def pindah(label):
    global no
    no = label
    
def ulang():
    print("")
    ulang = input("Apakah anda ingin kembali ke menu? y/n : ")
    if ulang == "y" or ulang == "Y" :
        pindah(0)
    elif ulang == "n" or ulang == "N" :
        print("")
        print("TERIMAKASIH, PROGRAM SELESAI")
        pindah(6)
    else :
        print("TERIMAKASIH, PROGRAM SELESAI")
        pindah(6)
        
def cek():
    global status, tunjangan1, jenis_kelamin, jumlah_tanggungan, tunjangan2
    if status == "Menikah":
        if jenis_kelamin == "L" or jenis_kelamin == "l":
            tunjangan2 = 200000
            tunjangan1 = int(jumlah_tanggungan) * 50000
        else:
            tunjangan2 = 150000
            tunjangan1 = int(jumlah_tanggungan) * 50000
    elif status == "Belum Menikah" :
        tunjangan2 = 0 
        tunjangan1 = 0
    else :
        print("Tulis Status Menikah/Belum Menikah")

def rumus():
    global gaji, tunjangan1, jumlah_tanggungan, tunjangan2, total_gaji
    total_gaji = int(gaji) + int(tunjangan2) + int(tunjangan1)

def cetak():
    print("")
    print("")
    print("===============================================================")
    print("                    PROGRAM GAJI KARYAWAN")
    print("===============================================================")
    print(" Hasil Inputan ")
    print("---------------------------------------------------------------")
    print("Nama Lengkap : ",nama)
    print("Jenis Kelamin : ",jenis_kelamin)
    print("Golongan : ",golongan)
    print("Status : ",status)
    print("Jumlah Tanggungan : ",jumlah_tanggungan)
    print("---------------------------------------------------------------")
    print("Total Gaji : Rp.",total_gaji)
    print("---------------------------------------------------------------")    
        
no = 0

while True :
    if no == 0 :
        print("")
        print("---------------------------------------------------------------")
        print("                         Gaji Karyawan")
        print("---------------------------------------------------------------")
        print(" Masukkan Inputan")
        print("---------------------------------------------------------------")
        nama = input("Nama Lengkap : ")
        jenis_kelamin = input("Jenis Kelamin (L/P) : ")
        golongan = input("Golongan : ")
        status = input("Status : ")
        jumlah_tanggungan = input("Jumlah Tanggungan : ")
        if golongan == "1" :
            pindah(1)
        elif golongan == "2" :
            pindah(2)
        elif golongan == "3" :
            pindah(3)
        else :
            pindah(0)
            
    elif no == 1 :
        gaji = 1000000
        cek()
        rumus()
        cetak()
        ulang()
    elif no == 2 :
        gaji = 2000000
        cek()
        rumus()
        cetak()
        ulang()
    elif no == 3 :
        gaji = 3000000
        cek()
        rumus()
        cetak()
        ulang()
    else :
        exit()
```
