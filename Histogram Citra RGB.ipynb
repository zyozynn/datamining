```python
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

cf = Image.open(r'C:\Users\asus\Documents\input\direktorat.jpg')
lbr, tgi = (cf.size)

bakir = []
bakig = []
bakib = []

for x in range (lbr):
    for y in range(tgi):
        r,g,b = cf.getpixel((x,y))
        bakir.append(r)
        bakig.append(g)
        bakib.append(b)
        
jb = 255
fig, axs = plt.subplots(1,3,figsize=(15,5))
# axs[0].imshow(cf)
axs[0].hist(bakir, jb, facecolor='red', alpha=0.5)
axs[1].hist(bakig, jb, facecolor='green', alpha=0.5)
axs[2].hist(bakib, jb, facecolor='blue', alpha=0.5)
plt.show()
```
